```python
from random import random
import string

interact = widgets.interactive.factory()
custom_interact_manual = interact.options(manual=True, manual_name="search")

class Catalog:
    def __init__(self):
        self.main_widget = widgets.VBox()
        self.widgets = []
#         self.catalog_items = []
        self.catalog_items = some_catalog_items
        
        
        self._w_search_input = widgets.Text()
        self._w_search_button = widgets.Button(
            description='search',
            icon='search'
        )
        def btn_onclick(*args):
            self._w_interactive_value.value = str(random())
            self.generate_catalog_item_list()

        self._w_search_button.on_click(btn_onclick)
        self._w_search_row = widgets.HBox([
            self._w_search_input,
            self._w_search_button
        ])
        
        self._w_interactive_value = widgets.ValueWidget()
        
        self.generate_catalog_item_list()
        
        def prepare_catalog_items(query):
            params = {'query': str(query)} if query else {}
            response = requests.get(
                f'http://localhost:8887/catalogue/items/',
                params=params,
                headers={'Authorization': f'Bearer {access_token}'})

            self.catalog_items = response.json()['items']
            print(str(query))
            self.generate_catalog_item_list()
            return self._w_catalog_items
            
        custom_interact_manual(
            prepare_catalog_items,
            query=self._w_search_input,
#             __options={
#                 'manual': True,
#                 'manual_name': 'search'
#             }
        )
#         {
#                 query': self._w_search_input,
#             }
        
        self.widgets = [
            self._w_search_row,
            self._w_catalog_items
        ]
        
        self.main_widget.children = self.widgets
#         display(self.main_widget)
        
    def generate_catalog_item_list(self):
        self._w_catalog_items = widgets.VBox(
            [
                SingleCatalogItemLabel(
                    catalog_item,
                    display_widget=False
                ).main_widget
                for catalog_item
                in self.catalog_items
            ]
        )

c = Catalog()
```
